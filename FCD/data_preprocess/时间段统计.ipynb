```python
import pandas as pd

import os

import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
```

```python
path = r'Excel\bh151\19\\'
for files in os.listdir(path):
    print(files)
    file = path + files
    temp = pd.read_excel(file,usecols = [2, 3, 4])
    df = df.append(temp)
```

```python
len(df)
```

```python
df = df.drop_duplicates() #去重
len(df)
```

```python
plates = []
diffs = []
```

```python
for name, group in df.groupby('Plate'):
#         print (name,group)
#         print(len(group))
        if len(group) > 1:
            diff = abs(group.iloc[1]['Time']-group.iloc[0]['Time'])
            if diff > 10000: #采样间隔大于10000秒的,就当是跨了一个周期，调整240000
                diff = 240000 - diff
                print('Adjusted period:', diff)
            plates.append(name)
            diffs.append(diff)
            print('Finished:', name, 'Period', diff)
```

```python
for name, group in df.groupby('Plate'):
#         print (name,group)
#         print(len(group))
        if len(group) > 1:
            time1 = group['Time'].as_matrix()
            
            time2 = np.append(time1,0)[1:]
            diff = abs(time1 - time2)
            
            diffmin = np.min(diff) #取最小的，当成是更新时间。由于已经去重了，所以按理说不应该有0
            if diffmin < 120: #超过120S的更新的，就当不是GPS的更新，而是车辆只在道路上短暂的停留或者是噪声
                plates.append(name)
                diffs.append(diffmin)
                print('Joined:', name, 'Period', diffmin)
            else:
                print('Excluded:', name, 'Period', diffmin)
```
